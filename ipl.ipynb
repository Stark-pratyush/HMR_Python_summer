```python
seasons=pd.unique(matches['season'])
```

```python
winner_in_each_season=[]
for season in seasons:
    matches_in_a_season=matches[matches['season']==season]
    winner_in_each_season.append(pd.value_counts(matches_in_a_season['winner']))
```

```python
for i in range(len(winner_in_each_season)):
    print(f'Season-{seasons[i]}')
    print(winner_in_each_season[i])
    print('\n')
```
